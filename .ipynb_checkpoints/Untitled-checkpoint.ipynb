```python
import pandas as pd
file = "Resources/purchase_data.csv"
main_db = pd.read_csv(file)
```

```python
main_db
```

```python
# Player Count & Total Purchasing Analysis
totalplayers = len(main_db.SN)
uniqueitems = len(main_db["Item Name"].value_counts())
averageprice = round(main_db.Price.mean(),2)
totalpurchases = len(main_db["Purchase ID"])
totalrevenue = main_db.Price.sum()
print(f"There are {totalplayers} total players")
print(f"There are {uniqueitems} unique items")
print(f"The average price is ${averageprice}")
print(f"There are {totalpurchases} total purchases")
print(f"The sum of all purchases is ${totalrevenue}")
```

```python
#Gender Demographics Counts & Purchase Counts
main_db["Gender"].value_counts()
```

```python
#Gender Demographics Percentage
main_db["Gender"].value_counts(normalize=True)
```

```python
#Average Purchase Price by Gender
main_db.groupby(by=["Gender"])["Price"].mean()
```

```python
#Total Purchase Value by Gneder
main_db.groupby(by=["Gender"])["Price"].sum()
```

```python
bins=[0,10,14,18,22,26,30,34,38,42,100]
labels = ["<10", "10-13", "14-17", "18-21", "22-25", "26-29", "30-33", "34-37", "38-41", "42+"]
main_db["Age Group"]=pd.cut(main_db.Age,bins=bins, labels=labels)
main_db
```

```python
#Purchase Count by Age Group
main_db.groupby(by=["Age Group"])["Age"].count()
```

```python
#Average Purchase Price by Age Group
main_db.groupby(by =["Age Group"])["Price"].mean()
```

```python
#Total Purchases by Age Group
main_db.groupby(by =["Age Group"])["Price"].sum()
```

```python
#Top Spenders
topspender = main_db.groupby(by="SN").sum()
topspender.sort_values(by="Price",inplace=True,ascending=False)
topspender.head(5)
```

```python
top5 = main_db.loc[(main_db["SN"] == "Lisosia93") | (main_db["SN"] == "Idastidru52") | (main_db["SN"] == "Chamjask73")
                   | (main_db["SN"] == "Iral74") | (main_db["SN"] == "Iskadarya95"),:]
top5count = top5.groupby(by="SN")["Purchase ID"].count()
top5count = top5count.to_frame()
top5average = top5.groupby(by="SN")["Price"].mean()
top5average = top5average.to_frame()
top5total = top5.groupby(by="SN")["Price"].sum()
top5total = top5total.to_frame()
```

```python
top5result = top5count.join(top5average, on="SN", how="outer")
top5result1 = top5result.join(top5total, on="SN", how='outer',lsuffix=' Average',rsuffix=' Total')
top5result1.rename(columns={"Purchase ID":"Total Purchases"},inplace=True)
top5result1
```

```python
#Most Popular Items
topitems = main_db.groupby(by="Item ID").count()
topitems.sort_values(by="Purchase ID",inplace=True,ascending=False)
topitems.head(5)
```

```python
top5items = main_db.loc[(main_db["Item ID"]==92) | (main_db["Item ID"]==178) | (main_db["Item ID"]==145) |
                        (main_db["Item ID"]==132) | (main_db["Item ID"]==108),:]
top5itemscount = top5items.groupby(by="Item Name")["Item ID"].count()
top5itemscount = top5itemscount.to_frame()
top5itemstotal = top5items.groupby(by="Item Name")["Price"].sum()
top5itemstotal = top5itemstotal.to_frame()
```

```python
top5itemfinal = top5itemscount.join(top5itemstotal, on="Item Name", how='outer')
top5itemfinal.rename(columns={"Item ID":"Items Sold", "Price":"Total Price"},inplace=True)
top5itemfinal["Item ID"]= [108, 145, 92, 178, 132]
top5itemfinal["Price"]=[3.53, 4.58, 4.88, 4.23, 3.19]
top5itemfinal
```

```python
#Most Profitable Items
topprofit = main_db.groupby(by="Item ID").sum()
topprofit.sort_values(by="Price",inplace=True,ascending=False)
topprofit.head(5)
```

```python
top5profit = main_db.loc[(main_db["Item ID"]==92) | (main_db["Item ID"]==178) | (main_db["Item ID"]==82) |
                        (main_db["Item ID"]==145) | (main_db["Item ID"]==103),:]
top5profitsum = top5profit.groupby(by="Item ID")["Price"].sum()
top5profitcount= top5profit.groupby(by="Item ID")["Purchase ID"].count()
top5profitsum=top5profitsum.to_frame()
top5profitcount=top5profitcount.to_frame()
```

```python
top5profittotal = top5profitsum.join(top5profitcount,on='Item ID', how='outer')
top5profittotal.rename(columns={"Purchase ID": "Total Items Sold","Price":"Total Value Sold"},inplace=True)
top5profittotal["Item Name"] = ["Final Critic", "Oathbreaker, Last Hope of the Breaking Storm", "Nirvana",
                                "Fiery Glass Crusader", "Singed Scalpel"]
top5profittotal["Price"] = [4.90, 4.19, 4.35, 4.58, 4.23]
top5profittotal
```
